**Data collected from 11 Letterboxd Users**

```python
import pandas as pd
df = pd.read_csv('users.csv')
df
```

**Data Cleaning**

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.rating.unique()
```

```python
df['rating'] = df['rating'].apply(lambda x:x/2 if x>10 else x)
```

```python
df.rating.unique()
```

```python
ratings = df.pivot_table(values='rating', index = 'username', columns = 'movie_name').fillna(0)
ratings
```

**Finding N Nearest Neighbours**

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=5)
model.fit(ratings)
```

```python
import joblib
joblib.dump(model, 'movie_recc.pkl')
```

**Sample Use case**

```python
input_ratings = [list(ratings.iloc[0, :].values)]
```

```python
_, neighbours = model.kneighbors(input_ratings)
```

```python
list(neighbours[0])
```

```python
n_ratings = ratings.iloc[list(neighbours[0])]
n_ratings
```

```python
n_ratings.mean().sort_values(ascending=False)
```
